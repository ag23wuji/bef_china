```r
clean_df <- function(df) {
  df %>%
    janitor::clean_names() %>%
    dplyr::mutate(across(
      where(is.character),
      ~ stringr::str_trim(tolower(.))
    ))
}
```

```r
library(dplyr)
library(stringr)

# Hilfsfunktion bleibt gleich
num_to_letters <- function(n) {
  # n should be numeric or coercible to numeric
  all_letters <- c(letters, paste0("a", letters))
  n_num <- as.numeric(n)
  # Return NA for invalid indices
  n_num[is.na(n_num) | n_num < 1 | n_num > length(all_letters)] <- NA
  return(all_letters[n_num])
}

# Für a_soils
a_soils_clean <- a_soils %>%
  mutate(
    ptag_clean = tolower(ptag),
    site_code = str_sub(ptag_clean, 2, 2),
    letter_num = str_sub(ptag_clean, 3, 4),
    plot_num = str_sub(ptag_clean, 5, 6),
    subplot_digit = str_sub(ptag_clean, 7, 7)
  ) %>%
  mutate(
    plot_letter = num_to_letters(letter_num),
    subplot_roman = case_when(
      subplot_digit == "0" ~ NA_character_,
      subplot_digit == "5" ~ "i",
      subplot_digit == "6" ~ "ii",
      subplot_digit == "7" ~ "iii",
      subplot_digit == "8" ~ "iv",
      TRUE ~ NA_character_
    ),
    join_id = if_else(
      is.na(subplot_roman),
      paste0(plot_letter, as.numeric(plot_num)),
      paste0(plot_letter, as.numeric(plot_num), "-", subplot_roman)
    )
  ) %>%
  select(ptag, sample_num, join_id) %>%
  distinct() %>%
  full_join(a_soils, by = c("ptag", "sample_num")) |> 
  mutate(join_id = paste0("a", join_id))  # <-- Korrigiert: paste0 statt paste

# Ausgabe der join_id-Spalten zur Kontrolle
a_soils_clean %>% select(ptag, join_id)
```

```r
soils_joined2 <- soils_joined |> 
  mutate(
    depth_lb = as.numeric(depth_lb),
    p_h_h2o = as.numeric(p_h_h2o),
    p_h_k_cl = as.numeric(p_h_k_cl),
    n_t = as.numeric(n_t),
    c_t = as.numeric(c_t), 
	c_n = as.numeric(c_n)
  ) |> 
 rename("plot" = join_id) |> 
 group_by(plot) |> 
 summarise_if(is.numeric, mean)
```

s_t wird entfernt wegen zu geringen Werten (hauptsächlich < 0.05). Alle Werte werden über die Tiefe hinweg gemittelt.

```r
soils_joined3 <- soils_joined2 |>
  group_by(plot) |>
  arrange(depth_lb) |>
  mutate(
    prev_depth = lag(depth_lb, default = 0),
    layer_thickness = depth_lb - prev_depth
  ) |>
  summarise(
    soil_h20 = sum(p_h_h2o * layer_thickness) / sum(layer_thickness),
    soil_ph_k_cl = sum(p_h_k_cl * layer_thickness) / sum(layer_thickness),
    soil_n = sum(n_t * layer_thickness) / sum(layer_thickness),
    soil_c = sum(c_t * layer_thickness) / sum(layer_thickness),
    soil_c_n = sum(c_n * layer_thickness) / sum(layer_thickness),
    .groups = "drop"
  )
```

```r
all_data_joined_lg_soil <- all_data_joined_lg |> 
 left_join(soils_joined3) 

all_data_joined_lg_soil |> 
 saveRDS(file = "datas_and_notebooks/final_datas/all_data_joined.rds")
```
