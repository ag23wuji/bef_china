```r
library(tidyverse)
library(readxl)

# Get all sheet names for each file
sheets_a <- excel_sheets("datas_and_notebooks/soil_data/SP6_Site_A_CNS_pH_depth_increments-2.xls")
sheets_b <- excel_sheets("datas_and_notebooks/soil_data/SP6_Site_B_CNS_pH_depth_increments.xls")

# Read all sheets into a named list of dataframes for each file
site_a_list <- setNames(
  lapply(sheets_a, function(s) read_excel("datas_and_notebooks/soil_data/SP6_Site_A_CNS_pH_depth_increments-2.xls", sheet = s)),
  sheets_a
)

site_b_list <- setNames(
  lapply(sheets_b, function(s) read_excel("datas_and_notebooks/soil_data/SP6_Site_B_CNS_pH_depth_increments.xls", sheet = s)),
  sheets_b
)
```

```r
clean_df <- function(df) {
  df %>%
    janitor::clean_names() %>%
    dplyr::mutate(across(
      where(is.character),
      ~ stringr::str_trim(tolower(.))
    ))
}
```

```r
a_soils <- site_a_list$'Raw data' |> 
 clean_df()
```

```r
b_soils <- site_b_list$`Raw data` |> 
 clean_df()
```
