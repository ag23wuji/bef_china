```r
library(dplyr)
library(stringr)

all_data2 <- all_data |>
  mutate(
    genus = case_when(
		
		species == "fraxinus insularis" ~ "fraxinus",
		species == "euonymus myrianthus" ~ "euonymus",
		species == "aralia chinensis" ~ "aralia",
		str_detect(species, "ilex ficoidea") ~ "ilex",
      str_detect(species, "adinandra millettii") ~ "adinandra",
      species == "alniphyllum fortunei" ~ "alniphyllum",
      species == "buddleja lindleyana" ~ "buddleja",
      str_detect(species, "camellia") ~ "camellia",
      species == "castanopsis sclerophylla" ~ "castanopsis",
      species == "crataegus cuneata" ~ "crataegus",
      species == "eurya muricata" ~ "eurya",
      species == "ilex pubescens" ~ "ilex", # Korrigiert von iley
      species %in% c("itea omeiensis", "itea sp.") ~ "itea",
      species == "lindera reflexa" ~ "lindera",
      species == "liquidambar formosana" ~ "liquidambar",
      str_detect(species, "litsea") ~ "litsea",
      species == "loropetalum chinense" ~ "loropetalum",
      species == "melia azedarach" ~ "melia",
      species == "mussaenda esquirolii" ~ "mussaenda", # Korrigiert von musseanda
      species == "phyllostachys glauca" ~ "phyllostachys",
      species == "premna microphylla" ~ "premna",
      species == "pterostyrax corymbosus" ~ "pterostyrax",
      str_detect(species, "quercus") ~ "quercus",
      str_detect(species, "recruit") ~ NA_character_,
      species == "rhaphiolepis indica" ~ "rhaphiolepis",
      species == "rhododendron simsii" ~ "rhododendron",
      species == "rhus chinensis" ~ "rhus",
      species == "rubus corchorifolius" ~ "rubus",
      species == "schoepfia jasminodora" ~ "schoepfia",
      species == "smilax china" ~ "smilax",
      str_detect(species, "symplocos") ~ "symplocos",
      species == "trema cannabina" ~ "trema",
      species == "viburnum setigerum" ~ "viburnum",
	  species == "musseanda esquirolii" ~ "musseanda",
      TRUE ~ genus
    ),
    family = case_when(
		genus == "musseanda" ~ "rubiaceae",
		genus == "fraxinus" ~ "oleaceae",
		genus == "euonymus" ~ "celastraceae",
		genus == "aralia" ~ "araliaceae",
		genus == "ilex" ~ "aquifoliaceae",
		species == "adinandra millettii" ~ "pentaphylacaceae",
      species == "alniphyllum fortunei" ~ "styracaceae",
      species == "buddleja lindleyana" ~ "scrophulariaceae",
      str_detect(species, "camellia") ~ "theaceae",
      species == "castanopsis sclerophylla" ~ "fagaceae",
      species == "crataegus cuneata" ~ "rosaceae",
      species == "eurya muricata" ~ "pentaphylacaceae",
      species == "ilex pubescens" ~ "aquifoliaceae",
      species %in% c("itea omeiensis", "itea sp.") ~ "iteaceae",
      species == "lindera reflexa" ~ "lauraceae",
      species == "liquidambar formosana" ~ "altingiaceae",
      str_detect(species, "litsea") ~ "lauraceae",
      species == "loropetalum chinense" ~ "hamamelidaceae",
      species == "melia azedarach" ~ "meliaceae",
      species == "mussaenda esquirolii" ~ "rubiaceae",
      species == "phoebe bournei" ~ "lauraceae",
      species == "phyllostachys glauca" ~ "poaceae",
      species == "premna microphylla" ~ "lamiaceae",
      species == "pterostyrax corymbosus" ~ "styracaceae",
      str_detect(species, "quercus") ~ "fagaceae",
      str_detect(species, "recruit") ~ NA_character_,
      species == "rhaphiolepis indica" ~ "rosaceae",
      species == "rhododendron simsii" ~ "ericaceae",
      species == "rhus chinensis" ~ "anacardiaceae",
      species == "rubus corchorifolius" ~ "rosaceae",
      species == "schoepfia jasminodora" ~ "schoepfiaceae",
      species == "smilax china" ~ "smilacaceae",
      str_detect(species, "symplocos") ~ "Symplocaceae",
      species == "trema cannabina" ~ "cannabaceae",
      species == "viburnum setigerum" ~ "adoxaceae",
      TRUE ~ family
    ),
    species = case_when(
      species == "knoopkruid" ~ "knoopkruid sp.",
		species == "styrax odoratissmus" ~ "styrax odoratissimus",
      TRUE ~ species
    )
  )
```

```r
all_data3 <- all_data2 |>
  mutate(
    growth_form = case_when(
      # Genus-basierte Bedingungen
      genus == "camellia" ~ "shrub/tree",
      genus == "castanopsis" ~ "tree",
      genus == "fraxinus" ~ "tree",
      genus == "ilex" ~ "shrub/tree",
      
      # Species-basierte Bedingungen
      species == "eurya muricata" ~ "shrub/tree",
      species == "itea omeiensis" ~ "shrub/tree",
      species == "lindera reflexa" ~ "shrub/tree",
      species == "liquidambar formosana" ~ "tree",
      species == "litsea cubeba" ~ "shrub/tree",
      species == "loropetalum chinense" ~ "shrub/tree",
      species == "melia azedarach" ~ "tree",
      species == "mussaenda esquirolii" ~ "shrub/tree",
      species == "phyllostachys glauca" ~ "shrub/tree",
      species == "premna microphylla" ~ "shrub/tree",
      species == "pterostyrax corymbosus" ~ "shrub/tree",
      species == "quercus sp." ~ "tree",
      species == "rhaphiolepis indica" ~ "shrub/tree",
      species == "rhododendron simsii" ~ "shrub",
      species == "rhus chinensis" ~ "shrub/tree",
      species == "rubus corchorifolius" ~ "shrub",
      species == "schoepfia jasminodora" ~ "shrub/tree",
      species == "smilax china" ~ "shrub/liana",
      species == "trema cannabina" ~ "shrub/tree",
      species == "adinandra millettii" ~ "shrub/tree",
	  species == "alniphyllum fortunei" ~ "tree",
		species == "buddleja lindleyana" ~ "shrub",
		species == "musseanda esquirolii" ~ "shrub/tree",
      # Fallback: Behalte die existierende growth_form für alle anderen
      TRUE ~ growth_form
    )
  )
```
